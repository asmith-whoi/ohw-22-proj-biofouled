## Quick Start
Hi Biofouled! Here is a quick start of working with the OOI example data sets. The annotations in these datasets have already been added as the "rollup_annotations_qc_flag" as part of a separate workflow.

```python
# Import libraries for data handling
import xarray as xr
import numpy as np
import pandas as pd
```

```python
# Import library for plotting the data
import matplotlib.pyplot as plt
```

#### Import the datasets

```python
example_data1 = xr.open_dataset("data/example_data1.nc")
example_data2 = xr.open_dataset("data/example_data2.nc")
```

```python
# Look at the first example dataset
example_data1
```

```python
# Plot the dataset with the annotations
# Set the figure size and axes
fig, ax = plt.subplots(figsize=(15, 9))

# Plot the oxygen data
ax.plot(example_data1["time"], example_data1["estimated_oxygen_concentration"], marker=".", linestyle="", color="tab:blue", label="Data")

# Find and plot just the annotations using "pandas-like" boolean masking
mask = example_data1["rollup_annotations_qc_results"] == 3
ax.plot(example_data1["time"][mask], example_data1["estimated_oxygen_concentration"][mask], marker=".", linestyle="", color="tab:red", label="Annotation")

# Add the lables, legends, grid, etc
ax.set_ylabel("estimated oxygen concentration", fontsize=14, weight="bold")
ax.grid()
ax.legend(fontsize=12)
ax.set_title("Global Station Papa Flanking Mooring A Dissolved Oxygen: Deployment 2", fontsize=14, weight="bold")

# Set the x-axis as a time axis
fig.autofmt_xdate()
```

```python
# Repeat the plotting for the second dataset
# Set the figure size and axes
fig, ax = plt.subplots(figsize=(15, 9))

# Plot the oxygen data
ax.plot(example_data2["time"], example_data2["estimated_oxygen_concentration"], marker=".", linestyle="", color="tab:blue", label="Data")

# Find and plot just the annotations using "pandas-like" boolean masking
mask = example_data2["rollup_annotations_qc_results"] == 3
ax.plot(example_data2["time"][mask], example_data2["estimated_oxygen_concentration"][mask], marker=".", linestyle="", color="tab:red", label="Annotation")

# Add the lables, legends, grid, etc
ax.set_ylabel("estimated oxygen concentration", fontsize=14, weight="bold")
ax.grid()
ax.legend(fontsize=12)
ax.set_title("Global Station Irminger Flanking Mooring A Dissolved Oxygen: Deployment 1", fontsize=14, weight="bold")

# Set the x-axis as a time axis
fig.autofmt_xdate()
```
